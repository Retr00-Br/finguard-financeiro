*   FinGuard é uma assistente em IA que será seu assistente educativo especializado no mercado financeiro brasileiro

```python
import speech_recognition as sr
import pyttsx3
import os

# Configuração da Persona do Agente
class FinGuardAI:
    def __init__(self):
        self.engine = pyttsx3.init()
        self.voices = self.engine.getProperty('voices')
        # Configura voz em português (ajuste o índice conforme seu SO)
        for voice in self.voices:
            if "brazil" in voice.name.lower():
                self.engine.setProperty('voice', voice.id)

        self.engine.setProperty('rate', 180) # Velocidade da fala

    def falar(self, texto):
        print(f"FinGuard: {texto}")
        self.engine.say(texto)
        self.engine.runAndWait()

    def ouvir(self):
        rec = sr.Recognizer()
        with sr.Microphone() as fala:
            # Validação de ruído
            rec.adjust_for_ambient_noise(fala, duration=1)
            print("Ouvindo (Pode falar)...")

            try:
                audio = rec.listen(fala, timeout=5)
                entrada = rec.recognize_google(audio, language="pt-BR")
                print(f"Usuário: {entrada}")
                return entrada.lower()
            except sr.UnknownValueError:
                return "não entendi"
            except sr.RequestError:
                return "erro de conexão"

# Instância do projeto
assistente = FinGuardAI()

def executar_assistente():
    assistente.falar("Olá, Pedro. Sou a FinGuard AI. Como posso ajudar em seus estudos financeiros hoje?")

    while True:
        comando = assistente.ouvir()

        # Lógica de integração com a Base de Conhecimento
        if "selic" in comando:
            assistente.falar("Consultando o Relatório Focus... A Taxa Selic é o principal instrumento de política monetária. No contexto atual das nossas fontes, ela dita o rendimento dos seus títulos pós-fixados.")

        elif "graham" in comando or "margem de segurança" in comando:
            assistente.falar("Segundo Benjamin Graham, a margem de segurança é a diferença entre o valor intrínseco e o preço de mercado. Isso protege o investidor contra erros de análise ou crises.")

        elif "sair" in comando:
            assistente.falar("Encerrando o atendimento. Lembre-se: o investimento inteligente exige disciplina. Até logo!")
            break

        elif comando == "não entendi":
            assistente.falar("Desculpe, não consegui captar. Poderia repetir?")

        else:
            assistente.falar("Pergunta recebida. Vou analisar isso com base nas nossas dez fontes do mercado financeiro.")

if __name__ == "__main__":
    executar_assistente()
```
